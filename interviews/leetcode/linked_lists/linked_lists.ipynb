```python

# A simple Python program to introduce a linked list

# Node class
class Node:

    # Function to initialise the node object
    def __init__(self, data):
        self.data = data  # Assign data
        self.next = None  # Initialize next as null


# Linked List class contains a Node object
class MyLinkedList:

    # Function to initialize head
    def __init__(self):
        self.head = None

    def get(self, index: int) -> int:
        """
        int get(int index) Get the value of the indexth
        node in the linked list. If the index is invalid,
        return -1.
        """
        # Set the cur to the head of the list.
        temp = self.head

        # Iterate through the list and return.
        for i in range(0,index):
            temp = temp.next

        return temp.data

    def print(self) -> None:
        """
        int get(int index) Get the value of the indexth
        node in the linked list. If the index is invalid,
        return -1.
        """
        # Iterate through the list and print.
        temp = self.head
        counter = 0
        while (temp):
            print (f'i: {counter} data: {temp.data}')
            counter += 1
            temp = temp.next

    def addAtHead(self, val: int) -> None:
        """
        Adding a node to the head of a linked list.
        Time Complexity: O(1)
        """

        new_node = Node(val)
        new_node.next = self.head
        self.head = new_node

    def addAtTail(self, val: int) -> None:
        """Adding a node to the head of a linked list."""
        temp = self.head

        # If the head if empty make new node the head.
        new_node =  Node(val)
        if temp == None:
            self.head == new_node
            return
        else:
            while (temp.next != None):
                temp = temp.next
            temp.next = new_node

    def addAtIndex(self, index: int, val: int) -> None:
        if index == 1:
            new_node = Node(val)
            new_node.next = self.head
            self.head = new_node
        i = 1
        n = self.head
        while i < index-1 and n is not None:
            n = n.next
            i = i+1
        if n is None:
            print("Index out of bound")
        else: 
            new_node = Node(val)
            new_node.next = n.next
            n.next = new_node

    def deleteAtIndex(self, index: int) -> None:

        temp = self.head
        for i in range(0,index):
            previous_temp = temp
            temp = temp.next
        previous_temp.next = temp.next

if __name__=='__main__':

    # Start with the empty list
    llist = MyLinkedList()

    llist.head = Node(11)
    second = Node(22)
    third = Node(33)
    fourth = Node(44)

    llist.head.next = second
    second.next = third
    third.next = fourth

    print('get value.')
    print(llist.get(2))

    llist.addAtHead(55)
    llist.addAtTail(66)
    llist.addAtIndex(1, 77)
    print('print linked list.')
    llist.print()
    print('testing delete')
    llist.deleteAtIndex(index=1)
```
